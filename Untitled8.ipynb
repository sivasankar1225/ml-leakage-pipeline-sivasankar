```python
import pandas as pd
import numpy as np

np.random.seed(42)

data = pd.DataFrame({
    'area_sqft': np.random.randint(500, 3000, 50),
    'num_bedrooms': np.random.randint(1, 5, 50),
    'age_years': np.random.randint(0, 30, 50)
})

# Target (price)
data['price_lakhs'] = (
    data['area_sqft'] * 0.05 +
    data['num_bedrooms'] * 10 -
    data['age_years'] * 0.3 +
    np.random.randn(50) * 5
)

data.head()
```

Build model

```python
from sklearn.linear_model import LinearRegression

X = data[['area_sqft', 'num_bedrooms', 'age_years']]
y = data['price_lakhs']

model = LinearRegression()
model.fit(X, y)

# Intercept
print("Intercept:", model.intercept_)

# Coefficients
for feature, coef in zip(X.columns, model.coef_):
    print(feature, ":", coef)

# Predictions
y_pred = model.predict(X)

# First 5 actual vs predicted
print(pd.DataFrame({
    'Actual': y[:5],
    'Predicted': y_pred[:5]
}))
```

Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)

print("MAE:", mae)
print("RMSE:", rmse)
print("R2:", r2)
```

# MAE: Average error in predictions
# RMSE: Penalizes large errors more than MAE
## R2: How well model explains variance (closer to 1 is better)bold text

```python
import matplotlib.pyplot as plt

residuals = y - y_pred

plt.hist(residuals, bins=10)
plt.title("Residual Distribution")
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.show()
```

# Residual = Actual - Predicted
# If histogram is centered around 0 → model is good
# If skewed → model has bias
